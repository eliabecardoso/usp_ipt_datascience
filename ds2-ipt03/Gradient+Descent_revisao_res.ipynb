Vamos minimizar o erro médio quadrático de uma regressão linear com 2 parâmetros :
t0 e t1.
Assim ye=t0+t1x
As amostras são:
x=1,y=10
x=2,y=4.5
x=3,y=2




Dadas as funções Erro Médio Quadrático (EMQ) e o gradiente de EMQ, com gradient descent, obtenha t0 e t1 (os parâmetros do modelo)

```python
def EMQ(t0,t1):
    return (1/3)*((t0+t1*1-10)**2+(t0+t1*2-4.5)**2+(t0+t1*3-2)**2)
def gEMQ(t0,t1):
    DPt0=(1/3)*(2*(t0+t1*1-10)+2*(t0+t1*2-4.5)+2*(t0+t1*3-2))
    DPt1=(1/3)*(2*(t0+t1*1-10)+2*(t0+t1*2-4.5)*2+2*(t0+t1*3-2)*3)
    return (DPt0,DPt1)
t0,t1=0.5,0.5
lr=0.1
for i in range(200):
    g=gEMQ(t0,t1)
    t0=t0-lr*g[0]
    t1=t1-lr*g[1]
    print(t0,t1,EMQ(t0,t1))
   
```

Vamos fazer o gráfico Modelo X Real

```python
x=[1,2,3]
y=[10,4.5,2]
ye=[]
for i in range(3):
    ye.append(t0+t1*x[i])
plt.plot(x,y,'bo')
plt.plot(x,ye,color='red')
plt.show()
```
