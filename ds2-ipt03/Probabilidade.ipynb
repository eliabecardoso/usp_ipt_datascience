#  Probabilidade

Definição: uma variável aleatória X representa um valor numérico associado a cada resultado de um experimento de probabilidade. A palavra aleatória indica que os valores assumidos por X são obtidos ao acaso.

Uma <b>variável aleatória discreta</b> pode assumir valores que podem ser contados. Exemplo: número de caras em 100 lançamentos de moedas. Números de pessoas que passam na Av. Paulista até 14h.



Uma <b>variável aleatória contínua</b> pode assumir qualquer valor numérico em um determinado intervalo ou série de intervalos. Isto é, uma variável aleatória contínua é uma variável para a qual o conjunto A é um conjunto infinito não enumerável. É uma variável que assume valores dentro de intervalos de números reais. Exemplo: altura de uma pessoa.

Uma <b>Distribuição de Probabilidade</b> é um modelo matemático que relaciona um valor de variável com sua probabilidade de ocorrência. No caso de uma variável discreta, a probabilidade dela assumir um valor X0 é P(x=Xo). Para variáveis contínuas, trabalhamos com a probabilidade de ser atingido um intervalo de valor. Exemplo P(x>X1).

Exemplo de <b>distribuição Discreta (com variável discreta) : Binomial</b>

Standardization de features: muitos algoritmos de ML e DL não funcionam adequadamente quando features diferentes são de ordens de grandeza muito distantes. Exemplo : 10^-3 e 10^5. Standardizar as features (não o target) é quase sempre uma boa medida para acelerar a convergência (ou mesmo para chegar à convergência) em alguns algoritmos. Na standardização, de cada valor é retirado a média e ele é dividido pelo desvio padrão. Após a standardização, a média da feature é zero e o desvio padrão é 1.
