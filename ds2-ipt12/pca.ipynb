# PCA


```python
import numpy as np
import pandas as pd
from sklearn import linear_model
from sklearn.metrics import accuracy_score
```

```python
df=pd.read_csv('c:\\dados\\iris.txt',names=['m1','m2','m3','m4','esp'])
df.head()
```

```python
m1=df['m1'].mean()
m2=df['m2'].mean()
m3=df['m3'].mean()
m4=df['m4'].mean()

df['m1']=df['m1']-m1
df['m2']=df['m2']-m2
df['m3']=df['m3']-m3
df['m4']=df['m4']-m4
```

```python
esp=df['esp'].copy()
esp=esp.map({'Iris-setosa':0,'Iris-virginica':1,'Iris-versicolor':0})
X=df.drop(columns=['esp']).copy()
```

```python
X[:5]
```

```python
C=np.cov(X.transpose())
C
```

```python
val,vec=np.linalg.eig(C)
val,vec
```

```python
np.dot(vec[:,0:2].T,X.T).T
```

```python
from sklearn.decomposition import PCA
p=PCA(2)
p.fit(X)
```

```python
print(p.components_)
print(p.explained_variance_)
B = p.transform(X)
print(B)
```

As features são agora ortogonais...nada fora da diagonal princial na variância

```python
np.cov(B.T)
```

Observe que a reconstrução é, obviamente, lossy.

```python
p.inverse_transform(B)
```
