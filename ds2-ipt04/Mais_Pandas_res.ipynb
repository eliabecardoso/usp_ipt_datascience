```python
krot=pd.read_csv('c:\\dados\\krot3.csv',sep=';',usecols=['Fechamento'],squeeze=True)
```

Retire as datas com cotação zero

```python
krot=krot[krot>0]
```

Vamos fazer uma média móvel com janela=5 dias e comparar com o gráfico normal

```python
mkrot=krot.rolling(5).mean()
```

```python
import matplotlib.pyplot as plt
plt.plot(krot,color='blue')
plt.plot(mkrot,color='black')
plt.show()
```
